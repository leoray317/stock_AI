```python
import pyhdfs
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
# 讀取balance_sheet資料

df_income_sheet = pd.read_csv('date_income_sheet.csv')
df_income_sheet_1101 = df_income_sheet[df_income_sheet['公司代號']=='1101']
df_income_sheet_1101.head()
```

```python
# 讀取balance_sheet資料

df_balance_sheet = pd.read_csv('date_balance_sheet.csv')
df_balance_sheet_1101 = df_balance_sheet[df_balance_sheet['公司代號']=='1101']
df_balance_sheet_1101
```

```python
# 讀取基本面資料

df_selected_stock = pd.read_csv('selected_stock.csv')
df_selected_stock['date'] = pd.to_datetime(df_selected_stock['date'], format='%Y/%m/%d').dt.strftime('%Y-%m-%d')
df_selected_stock.head()
```

```python
print(df_selected_stock.shape)
print(df_balance_sheet_1101.shape)
print(df_income_sheet_1101.shape)
```

```python
pd.merge(df_selected_stock,df_balance_sheet_1101, how="inner",on='date')
```

```python
# 合併所有資料

pd.concat([df_selected_stock,df_balance_sheet_1101], axis=1,join="inner")
```
